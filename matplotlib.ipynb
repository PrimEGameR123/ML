<h1> What is matplotlib? </h1>
<font size = "+1"> matplotlib is a Python library specializing in the development of two-dimensional charts (including 3D charts).<br>
    Among all the features that have made it the most used tool in the graphical representation of data, there are a few that stand out:<br>
<br>• Extreme simplicity in its use
<br>• Gradual development and interactive data visualization
<br>• Greater control over graphic elements
<br>• Export to many formats, such as PNG, PDF, SVG, and EPS<br><br>
    In fact, with regard to data analysis, matplotlib normally cooperates with a set
of other libraries such as NumPy and pandas, but many other libraries can be integrated
without any problem

<font size = "+1"> The architecture of matplotlib is logically structured into three layers, which are placed at three different levels. <br>The three layers are as follows:<br>
• Scripting<br>
• Artist<br>
• Backend<br>
    <center>  </center>
    The communication is unidirectional, that is, each layer can communicate with the underlying layer, while the lower layers cannot communicate with the top ones.

<h2> 1. Backend layer - </h2>
<font size = "+1"> This layer contains the matplotlib APIs, a set of classes that play the role of implementation of the graphic elements at a low level.
    <ul>• FigureCanvas is the object that embodies the concept of drawing area.</ul>
    <ul>• Renderer is the object that draws on FigureCanvas.</ul>
<ul>• Event is the object that handles user inputs (keyboard and mouse events).</ul>
    It is similar to how we do drawing on paper. Consider you want to draw a painting. You get a blank paper (FigureCanvas) and a brush (Renderer). You ask your friend what to draw (Event).

<h2> 2. Artist Layer </h2>
<font size = "+1">All the elements that make up a chart, such as the title, axis labels, markers, etc., are instances of the Artist object.<br>

<font size = "+1"> There are two Artist classes: 
    <ul> • primitive -  Line, Circle, Text</ul>
    <ul> • composite - Figure, Axes (can be refer as plot or chart), axis (numerical values to be represented on Axes)</ul>
   <font size = "+1"> 1.  The primitive artists are individual objects that constitute the basic elements to form a graphical representation in a plot, for example a Line2D, or as a geometric figure such as a Rectangle or Circle, or even pieces of text.<br>
  <font size = "+1">  2. The composite artists are those graphic elements present in a chart that are composed of several base elements, namely, the primitive
artists. Composite artists are for example the Axis, Ticks, Axes, and Figures.

 <font size = "+0.5"> Figure is the main Artist object that holds everything together. 

<h2>3. Scripting Layer </h2>
<font size = "+1">  1. For the analysis and visualization of data, the scripting layer is best. This layer consists of an interface called pyplot.<br>
2. The pyplot module is a collection of command-style functions that allow you to use matplotlib.<br>
3. Each pyplot function will operate or make some changes to the Figure object, for example, the creation of the Figure itself, the creation of a plotting area, representation of a line, decoration of the plot with a label, etc.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# , you simply use the plot() function to pass the values to be plotted.
plt.plot([1,2,3,4],[1,2,3,4])
```

```python
# Let's plot a sine wave
time = np.arange(0,10,0.1)
amplitude = np.sin(time)
plt.plot(time, amplitude)
```

<font size = "+0.5"> <u>a default configuration of the plt.plot() function: </u><br>
• The size of the axes matches perfectly with the range of the input data.<br>
• There is neither a title nor axis labels.<br>
• There is no legend.<br>
• A blue line connecting the points is drawn

```python
# adding multiple styiles to plot and modifying the defaults
plt.xlabel("Time")
plt.ylabel("Amplitude = sin(Time)")
plt.title("Sine Wave")
plt.grid(True, which='both')
plt.axhline(y = 0, color = 'r', ls = '--', linewidth = 2)
plt.plot(time, amplitude,'g*')
plt.legend(["y = 0","sine"])
```

<h2> Working with Multiple Figures and Axes </h2>
<font size = "+1"> Actually, matplotlib allows you to manage multiple figures simultaneously, and within each figure, it offers the ability to view different plots defined as subplots.<br>
    The subplot() function, subdivides the figure in different drawing areas, and focus the commands on a specific subplot.<br>
    subplot( m , n , p ) divides the current figure into an m -by- n grid and creates axes in the position specified by p . 

```python
t = np.arange(0,5,0.1)
y1 = np.sin(np.pi*t)
y2 = np.cos(np.pi*t)
```

```python
plt.subplot(211)
plt.plot(t,y1,'b-.')
plt.subplot(212)
plt.plot(t,y2,'r--')
plt.show()
```

<h3> 4. Pie Chart </h3>

```python
labels = ['Nokia','Samsung','Apple','Lumia','Google']
values = [10,30,45,15,100]
plt.pie(values, labels=labels, autopct='%1.1f')
plt.show()
```

<font size = "+0.5"> for example, you would highlight the slice referring to Nokia. In order to do this, there is a special kwarg named explode. It is nothing but a sequence of float values of 0 or 1, where 1 corresponds to the fully extended slice and 0 corresponds to slices completely in the pie.

```python
explod = [0.4,0,0.1,0,0]
```

```python
plt.pie(values, labels=labels, autopct='%1.1f',explode =explod, shadow=True)
plt.show()
```
